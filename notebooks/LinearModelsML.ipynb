# Problem 1
(*3 points* )

To prepare for a test, FCS students eat candies and solve problems. The number of solved problems $y$ depends on the number of eaten candies $x$. If a student has not eaten any candy, they do not want to solve problems. Therefore, for describing the dependency of the number of solved tasks on the number of eaten candies, a linear model with one feature and no constant is used: $y_i = w \cdot x_i.$ Find an analytic expression of estimates of the parameter $w$ by minimizing the following loss function:

$$Q(w) = \frac{1}{\ell} \sum_{i=1}^{\ell} (y_i - w x_i)^2.$$

## Problem 1

We need to find $w$ that minimizes $Q(w) = \frac{1}{\ell} \sum_{i=1}^{\ell} (y_i - w x_i)^2$.

We take the derivative of $Q(w)$ with respect to $w$:

$$\frac{dQ}{dw} = \frac{1}{\ell} \sum_{i=1}^{\ell} 2(y_i - w x_i)(-x_i) = -\frac{2}{\ell} \sum_{i=1}^{\ell} x_i(y_i - w x_i).$$

Now we set the derivative equal to zero:

$$-\frac{2}{\ell} \sum_{i=1}^{\ell} x_i(y_i - w x_i) = 0$$

$$\sum_{i=1}^{\ell} x_i y_i - w \sum_{i=1}^{\ell} x_i^2 = 0$$

$$w \sum_{i=1}^{\ell} x_i^2 = \sum_{i=1}^{\ell} x_i y_i$$

And we solve for $w$:

$$w = \frac{\sum_{i=1}^{\ell} x_i y_i}{\sum_{i=1}^{\ell} x_i^2}$$

To make sure this is a minimum and not a maximum, we check the second derivative: $\frac{d^2 Q}{dw^2} = \frac{2}{\ell}\sum_{i=1}^{\ell} x_i^2 > 0$: as long as not all $x_i = 0$, so this is indeed a minimum.

# Problem 2

James measured the weight of three Pokemons, $y_1=6$, $y_2=6$, $y_3=10$ (in kg). Now James wants to predict weight for the next Pokemon using a constant model  $y_i = w$. For estimating the parameter $w$ he uses the following target function:

$$\frac{1}{\ell}\sum_{i=1}^{\ell} (y_i - w)^2 + \lambda w^2.$$

1. Find an optimal $w$ for an arbitrary $\lambda$ (*2 points* )
1. Find an optimal $\lambda$ using LOOCV (leave one out cross-validation) (*2 points* )
> LOOCV algorithm utilizes the following logic: On the first step, we evaluate the model on the entire sample without the first observation, then test it on the first observation. On the second step, we evaluate the model on the entire sample without the second observation, and now test it on the second observation. The algorithm is performed $\ell$ times. In order to find $\lambda_{CV}$ we minimize the average error made on the test samples.

1. Find an optimal value of $w$ for $\lambda_{CV}$, found with the algorithm above (*2 points* )
1. Derive an analytical expression for $\lambda_{CV}$ for an arbitrary number of observations (*2 points* )

## Solution to Problem 2.1

We need to find the optimal $w$ for an arbitrary $\lambda$

We have $y_1 = 6$, $y_2 = 6$, $y_3 = 10$, and $\ell = 3$. The objective function is

$$L(w) = \frac{1}{3}\sum_{i=1}^{3}(y_i - w)^2 + \lambda w^2$$

Taking the derivative and setting it to zero

$$\frac{dL}{dw} = \frac{1}{3}\sum_{i=1}^{3} 2(w - y_i) + 2\lambda w = 0$$

$$\frac{2}{3}\left(3w - \sum_{i=1}^{3} y_i\right) + 2\lambda w = 0$$

Since $\sum y_i = 6 + 6 + 10 = 22$

$$\frac{2}{3}(3w - 22) + 2\lambda w = 0$$

$$2w - \frac{44}{3} + 2\lambda w = 0$$

$$w(2 + 2\lambda) = \frac{44}{3}$$

$$w(\lambda) = \frac{22}{3(1 + \lambda)}$$

Note: for $\lambda = 0$ we get $w = \frac{22}{3} \approx 7.33$, which is just the sample mean, as expected

## Solution to Problem 2.2

We need to find the optimal $\lambda$ using LOOCV.

For LOOCV, at each fold $k \in \{1, 2, 3\}$ we leave out observation $y_k$, train on the remaining $\ell - 1 = 2$ observations, and test on $y_k$.

Training on fold $k$ (leaving out $y_k$): we minimize over the remaining 2 observations:

$$L_{-k}(w) = \frac{1}{2}\sum_{i \neq k}(y_i - w)^2 + \lambda w^2$$

Taking derivative and setting to zero

$$\frac{1}{2}\cdot 2\sum_{i \neq k}(w - y_i) + 2\lambda w = 0$$

$$2w - \sum_{i \neq k} y_i + 2\lambda w = 0$$

$$w_{-k} = \frac{\sum_{i \neq k} y_i}{2(1 + \lambda)}$$

Computing for each fold

For $k=1$: $\sum_{i \neq 1} y_i = y_2 + y_3 = 6 + 10 = 16$, so $w_{-1} = \frac{16}{2(1+\lambda)} = \frac{8}{1+\lambda}$

For $k=2$: $\sum_{i \neq 2} y_i = y_1 + y_3 = 6 + 10 = 16$, so $w_{-2} = \frac{8}{1+\lambda}$

For $k=3$: $\sum_{i \neq 3} y_i = y_1 + y_2 = 6 + 6 = 12$, so $w_{-3} = \frac{6}{1+\lambda}$

LOOCV error

$$CV(\lambda) = \frac{1}{3}\sum_{k=1}^{3}(y_k - w_{-k})^2 = \frac{1}{3}\left[\left(6 - \frac{8}{1+\lambda}\right)^2 + \left(6 - \frac{8}{1+\lambda}\right)^2 + \left(10 - \frac{6}{1+\lambda}\right)^2\right]$$

Let $t = \frac{1}{1+\lambda}$, where $\lambda \geq 0$ means $t \in (0, 1]$.

$$CV(t) = \frac{1}{3}\left[2(6 - 8t)^2 + (10 - 6t)^2\right]$$

Expanding

$$= \frac{1}{3}\left[2(36 - 96t + 64t^2) + (100 - 120t + 36t^2)\right]$$

$$= \frac{1}{3}\left[72 - 192t + 128t^2 + 100 - 120t + 36t^2\right]$$

$$= \frac{1}{3}\left[164t^2 - 312t + 172\right]$$

Taking the derivative with respect to $t$ and setting to zero:

$$\frac{dCV}{dt} = \frac{1}{3}(328t - 312) = 0 \implies t = \frac{312}{328} = \frac{39}{41}$$

Second derivative: $\frac{d^2 CV}{dt^2} = \frac{328}{3} > 0$ — this is a minimum.

Now recover $\lambda$:

$$\frac{1}{1+\lambda} = \frac{39}{41} \implies 1 + \lambda = \frac{41}{39} \implies \lambda_{CV} = \frac{2}{39}$$

## Solution to Problem 2.3

We need to find the optimal value of $w$ for $\lambda_{CV} = \frac{2}{39}$.

Using the formula from 2.1:

$$w = \frac{22}{3(1 + \lambda_{CV})} = \frac{22}{3\left(1 + \frac{2}{39}\right)} = \frac{22}{3 \cdot \frac{41}{39}} = \frac{22 \cdot 39}{3 \cdot 41} = \frac{858}{123} = \frac{286}{41}$$

$$w = \frac{286}{41} \approx 6.976$$

## Solution to Problem 2.4

We need to derive an analytical expression for $\lambda_{CV}$ for an arbitrary number of observations.

Let $y_1, \ldots, y_\ell$ be the observations and $S = \sum_{i=1}^{\ell} y_i$.

First, we find $w_{-k}$.

Training on fold $k$ (leaving out $y_k$), the objective is:

$$L_{-k}(w) = \frac{1}{\ell - 1}\sum_{i \neq k}(y_i - w)^2 + \lambda w^2$$

Setting the derivative to zero:

$$\frac{2}{\ell - 1}\left((\ell - 1)w - \sum_{i \neq k} y_i\right) + 2\lambda w = 0$$

$$2w - \frac{2(S - y_k)}{\ell - 1} + 2\lambda w = 0$$

$$w_{-k} = \frac{S - y_k}{(\ell - 1)(1 + \lambda)}$$

Now we write the LOOCV error:

$$CV(\lambda) = \frac{1}{\ell}\sum_{k=1}^{\ell}\left(y_k - \frac{S - y_k}{(\ell - 1)(1 + \lambda)}\right)^2$$

Let $t = \frac{1}{(\ell - 1)(1 + \lambda)}$. Then:

$$CV(t) = \frac{1}{\ell}\sum_{k=1}^{\ell}\left(y_k - (S - y_k)\cdot t\right)^2$$

Now we minimize over $t$:

$$\frac{dCV}{dt} = \frac{2}{\ell}\sum_{k=1}^{\ell}\left(y_k - (S - y_k)t\right)\cdot\left(-(S - y_k)\right) = 0$$

$$\sum_{k=1}^{\ell} (S - y_k)\left(y_k - (S - y_k)t\right) = 0$$

$$\sum_{k=1}^{\ell} y_k(S - y_k) = t\sum_{k=1}^{\ell}(S - y_k)^2$$

$$t^* = \frac{\sum_{k=1}^{\ell} y_k(S - y_k)}{\sum_{k=1}^{\ell}(S - y_k)^2}$$

Now we simplify the sums.

Denote $R = \sum_{k=1}^{\ell} y_k^2$.

Numerator:
$$\sum_{k=1}^{\ell} y_k(S - y_k) = S\sum_{k=1}^{\ell} y_k - \sum_{k=1}^{\ell} y_k^2 = S^2 - R$$

Denominator:
$$\sum_{k=1}^{\ell}(S - y_k)^2 = \sum_{k=1}^{\ell}(S^2 - 2Sy_k + y_k^2) = \ell S^2 - 2S^2 + R = (\ell - 2)S^2 + R$$

So:

$$t^* = \frac{S^2 - R}{(\ell - 2)S^2 + R}$$

Finally, we recover $\lambda$.

Since $t = \frac{1}{(\ell - 1)(1 + \lambda)}$, we have:

$$(\ell - 1)(1 + \lambda) = \frac{1}{t^*} = \frac{(\ell - 2)S^2 + R}{S^2 - R}$$

$$1 + \lambda = \frac{(\ell - 2)S^2 + R}{(\ell - 1)(S^2 - R)}$$

$$\lambda_{CV} = \frac{(\ell - 2)S^2 + R}{(\ell - 1)(S^2 - R)} - 1 = \frac{(\ell - 2)S^2 + R - (\ell - 1)S^2 + (\ell - 1)R}{(\ell - 1)(S^2 - R)} = \frac{\ell R - S^2}{(\ell - 1)(S^2 - R)}$$

where $S = \sum_{i=1}^{\ell} y_i$ and $R = \sum_{i=1}^{\ell} y_i^2$.

Verification with Problem 2.2 data: $S = 22$, $R = 36 + 36 + 100 = 172$, $\ell = 3$:

$$\lambda_{CV} = \frac{3 \cdot 172 - 484}{2 \cdot (484 - 172)} = \frac{516 - 484}{2 \cdot 312} = \frac{32}{624} = \frac{2}{39}$$

This matches our earlier result.

## Answers to self-check questions

### 1. Hyperparameter vs parameter

A parameter is something the model learns from data during training. For example, the weight $w$ in linear regression - the model finds it by minimizing the loss function. A hyperparameter is something we choose ourselves before training starts. For example, the regularization coefficient $\lambda$ in ridge regression. We usually pick hyperparameters using cross-validation.

### 2. Cross-validation

Cross-validation is a way to evaluate how well a model generalizes. We split the data into $K$ parts (folds). Then we train the model on $K-1$ folds and test it on the remaining one. We repeat this $K$ times so every fold gets to be the test set once. The final score is the average error across all folds.

It is better than a simple train-test split because it uses all the data for both training and testing. The result does not depend on one random split, so the estimate of the error is more reliable.

### 3. Categorical features and one-hot encoding

If we encode categories with numbers ("red" = 1, "blue" = 2, "green" = 3), the model thinks there is an order and that the distances between categories matter. But for nominal categories like colors this makes no sense.

One-hot encoding solves this. We create a separate binary column for each category. For example, with 3 colors we get 3 columns, and for each row only one of them is 1, the rest are 0. This way the model does not assume any ordering between categories.

### 4. Sensitivity of MSE to outliers

MSE uses squared errors: $\text{MSE} = \frac{1}{n}\sum(y_i - \hat{y}_i)^2$. Because we square the errors, big mistakes get much bigger. For example, if one point has an error of 10, it adds $10^2 = 100$ to the total, while a normal point with error 1 adds only $1^2 = 1$. So one outlier can have a huge effect on the result. MAE (mean absolute error) does not square the errors and is more robust to outliers.

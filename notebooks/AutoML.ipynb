3. Use splines on the number columns. Some number columns have a curved relationship with the targets. For example, when horsepower goes up, price goes up faster. 4. One-hot encoding for category columns. This turns each category into a 0/1 column then the model can use it. We also tell the encoder to ignore unknown categories and that helped the model not to break later.

6. Add Nystrom RBF features. It creates new features that help the linear model find more complex patterns. It works like SVM but is faster.

We also use log on all three targets to make the numbers easier to learn. This helps most for price because it has a long tail with some very expensive cars.

How we checked if it works

What we did

We chose the regularization force (alpha = 1.0), tried several different values and saved what works best. This gave us the strength of a complex model that complies with the rules.

2. Why it works. Nystrom RBF features and Ridge gives almost the same result as RBF-kernel SVM, but works faster when using big data. Thus, we get a strong model without using slow methods.

3. Log transform for all three targets. All targets are positive numbers, and price has a long tail, because most cars are cheap and some are very expensive. If we use log on the targets we get the values more even and it helps the model. The biggest improvement is on price.

4. One model per target. Each our target has its own model in case of difference in scale and shape. This is much better than sharing one model for all others.

5. Clip price at zero. The model can gives a negative number for price, but as we know the price cannot be negative, so we set negative predictions to zero.

How we checked if it works

We looked at the score on a special validation split, R² for each target. Final score on Kaggle: about 0.87 and the baseline was about 0.83, so our model is definitely better.

What we also tried

2. More Nystrom features (500), gave a small score improvement but went over the time limit. We kept 400 as a good balance.
3. Bigger sample (150000 or 200000 rows), small score improvement but too slow.
4. Splines with no Nystrom and many knots, simpler and safer, but the score was lower (about 0.86).
5. Polynomial interactions instead of Nystrom, also slower and a bit worse.

Below is a baseline model that produces the result on Kaggle leaderboard (LB).

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer, TransformedTargetRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, SplineTransformer
from sklearn.linear_model import Ridge
from sklearn.kernel_approximation import Nystroem

num_cols = tX.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in tX.columns if c not in num_cols]

N_SUB = 100_000
rng = np.random.RandomState(0)
idx = rng.choice(len(tX), size=N_SUB, replace=False)
tX_s = tX.iloc[idx].reset_index(drop=True)
tY_s = tY.iloc[idx].reset_index(drop=True)

numeric_pipe = Pipeline([
    ('scale',  StandardScaler()),
    ('spline', SplineTransformer(n_knots=6, degree=3, include_bias=False)),
])
categorical_pipe = Pipeline([
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False)),
])
pre = ColumnTransformer([
    ('num', numeric_pipe,     num_cols),
    ('cat', categorical_pipe, cat_cols),
], remainder='drop')

features = Pipeline([
    ('pre', pre),
    ('nys', Nystroem(kernel='rbf', gamma=0.1, n_components=400, random_state=0)),
])

def make_model_for_target(target_name):
    base = Pipeline([
        ('feat', features),
        ('reg',  Ridge(alpha=1.0, random_state=0)),
    ])
    return TransformedTargetRegressor(
        regressor=base, func=np.log1p, inverse_func=np.expm1
    )

models = {}
for t in YCols:
    m_t = make_model_for_target(t)
    m_t.fit(tX_s, tY_s[t].values)
    print(f'in-sample R^2 [{t}] = {m_t.score(tX_s, tY_s[t].values):.4f}')
    models[t] = m_t
```

```python
pY = pd.DataFrame(index=vX.index, columns=YCols, dtype=float)
for t in YCols:
    pY[t] = models[t].predict(vX)
pY['price'] = pY['price'].clip(lower=0)
ToCSV(pY, 'Auto_baseline')
print(pY.head())
print('Saved Auto_baseline.csv')
```
